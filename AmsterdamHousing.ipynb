```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# ... (rest of your code) ...
amsterdam_file_path = '/content/HousingPrices-Amsterdam-August-2021.csv'
amsterdam_data = pd.read_csv(amsterdam_file_path)
amsterdam_data.columns

# Remove rows with missing 'Price' (target variable) before imputing
amsterdam_data = amsterdam_data.dropna(subset=['Price'])

amsterdam_features = ['Area', 'Room']

X = amsterdam_data[amsterdam_features]
X.describe()

# Create an imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on your feature data and transform it
X_imputed = imputer.fit_transform(X)  # Use the updated X and store in a new variable

# Convert the imputed NumPy array back to a Pandas DataFrame
X = pd.DataFrame(X_imputed, columns=amsterdam_features, index=X.index)

y = amsterdam_data.Price  # Update y after removing rows

# Define and fit the model
amsterdam_model = DecisionTreeRegressor(random_state=1)
amsterdam_model.fit(X, y)


def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# # the random_state argument guarantees we get the same split every time we
# # run this script.
# train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
# # Define model
# amsterdam_model = DecisionTreeRegressor()
# # Fit model
# amsterdam_model.fit(train_X, train_y)

# # get predicted prices on validation data
# val_predictions = amsterdam_model.predict(val_X)
# print(mean_absolute_error(val_y, val_predictions))

max_leaf_nodes = [5, 25, 50, 100, 250, 500]
my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

```
